```python
from matplotlib import pyplot as plt
import numpy as np
import os
%matplotlib notebook
```

```python
e = np.arange(-1.5,1.50,.1)
def a_domains(x):
    return np.where(x<-1,0,np.where(x<1,100*(1+x)/2,100))
    
plt.figure(figsize=(6,3))
plt.plot(e,a_domains(e))
plt.xlabel(r'$\epsilon/\epsilon_s(T)$')
plt.ylabel('Proportion of "a" domains (%)')
```
