```python
Mdata = pd.read_csv('Y05TmVO4-LS5250-MS180521_MvsT_20180521.dat', header=26)
Mdata
```

```python
M = Mdata['Moment (emu)']
Merr = Mdata['M. Std. Err. (emu)']
T = Mdata['Temperature (K)']
H = Mdata['Magnetic Field (Oe)']
```

```python
def m_plot_settings(axis):
    axis.set_ylabel(r'$M$ (emu)')
    axis.set_xlabel(r'$T$ (K)')
    axis.locator_params(nbins=6)
```

```python
fig, ax = plt.subplots()
plt.errorbar(T, M, yerr=Merr, marker='.', ms=10, lw=0, elinewidth=2)
m_plot_settings(ax)
ax.ticklabel_format(axis='y', scilimits=(-3,6))
ax.set_title(f'Magnetization of Tm$_{{0.95}}$Y$_{{0.05}}$VO$_4$ at $H$ = {H[0]:.0f} Oe')#
# ax.set_xlim(left=1.8)
```
